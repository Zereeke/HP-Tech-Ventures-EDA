```python
# Function to compare original and imputed data distributions
def compare_distributions(original, imputed, column):
    plt.figure(figsize=(10, 6))
    sns.kdeplot(original[column].dropna(), label='Original')
    sns.kdeplot(imputed[column], label='Imputed')
    plt.title(f'Distribution Comparison - {column}')
    plt.legend()
    plt.show()

# Compare distributions for numerical columns
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    compare_distributions(df, df_mice_imputed, col)
```

```python
# Function to perform a simple analysis (e.g., correlation with target variable)
def simple_analysis(data, target_column):
    return round(data.corr()[target_column].sort_values(ascending=False), 1)

# Compare results across different imputation methods
results = {
    'Original': simple_analysis(df, 'Last Funding Amount Currency (in USD)'),
    'Mean Imputed': simple_analysis(df_mean_imputed, 'Last Funding Amount Currency (in USD)'),
    'Median Imputed': simple_analysis(df_median_imputed, 'Last Funding Amount Currency (in USD)'),
    'Mode Imputed': simple_analysis(df_mode_imputed, 'Last Funding Amount Currency (in USD)'),
    'KNN Imputed': simple_analysis(df_knn_imputed, 'Last Funding Amount Currency (in USD)'),
    'MICE Imputed': simple_analysis(df_mice_imputed, 'Last Funding Amount Currency (in USD)'),
    
}

# Print results
for method, result in results.items():
    print(f"\n{method}:")
    print(result)
```

```python
from sklearn.cluster import KMeans
```

```python
def load_glove_embeddings(file_path):
    embeddings_index = {}
    with open(file_path, 'r', encoding = 'utf-8') as f:
        for line in f:
            values = line.split()
            word = values[0]
            try:
                coefs = np.asarray(values[1:], dtype = 'float32')
                embeddings_index[word] = coefs
            except ValueError:
                continue
    return embeddings_index

glove_file_path = '/Users/somipatel/Desktop/HP Business Analytics/glove.6B/glove.840B.300d.txt'
```

```python
embeddings_index = load_glove_embeddings(glove_file_path)
```

```python
def get_average_embedding(text, embeddings_index, dim = 100):
    words = text.split()
    valid_embedding = [embeddings_index.get(word, np.zeros(dim)) for word in words]
    if valid_embedding:
        embedding = np.mean(valid_embedding, axis=0)
    else:
        embedding = np.zeros(dim)
    return embedding
```

```python
embeddings = np.array([get_average_embedding(industry, embeddings_index, 300) for industry in all_industries])
```

```python
num_clusters = 14
kmeans = KMeans(n_clusters = num_clusters, random_state = 42)
kmeans.fit(embeddings)
```

```python
industry_clusters = {industry: cluster for industry, cluster in zip(all_industries, kmeans.labels_)}
```

```python
# Importing necessary libraries
from sklearn import preprocessing  # Library for data preprocessing
from sklearn.model_selection import train_test_split  # Library for splitting data into training and testing sets
import statsmodels.api as sm  # Library for statistical models

# Selecting numerical columns from the DataFrame
num_columns = final.select_dtypes(exclude=['object', 'datetime64[ns]']).columns

# Creating a new DataFrame 'final1' with only numerical columns
final1 = pd.DataFrame(final, columns=num_columns)

# Converting columns related to industry (boolean) to integers
for col in industry_columns:
    final1[col] = final1[col].astype(int)

# Adding a squared term of 'Last Funding Amount Currency (in USD)' to capture non-linear relationships
#final1['Last Funding Amount Currency (in USD)^2'] = final1['Last Funding Amount Currency (in USD)'] ** 2

# Copying industry columns into a new list to avoid modifying the original list
feature_list = industry_columns.copy()
feature_list.extend(['Last Funding Amount Currency (in USD)',
                     'Number of Employees',
                     'Number of Funding Rounds',
                     'Number of Founders',
                     'Estimated Revenue Range'
                    ])

# Selecting features based on the feature list
features = final1[feature_list]

# Setting the target variable (y) as 'Total Funding Amount Currency (in USD)'
y = final1['Total Funding Amount Currency (in USD)']


# Splitting the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(features, y, test_size=0.2, random_state=42)

# Adding a constant term to the training features (necessary for statsmodels)
X_with_const_train = sm.add_constant(X_train)

# Fitting an Ordinary Least Squares (OLS) regression model on the training data
# OLS: A method for estimating the parameters in a linear regression model by minimizing the sum of squared residuals
model_sm_train = sm.OLS(y_train, X_with_const_train).fit()

# Printing the summary of the OLS regression model on training data
print(model_sm_train.summary())

# Adding a constant term to the testing features
X_with_const_test = sm.add_constant(X_test)

# Fitting an OLS regression model on the testing data
model_sm_test = sm.OLS(y_test, X_with_const_test).fit()

# Printing the summary of the OLS regression model on testing data
print(model_sm_test.summary())
```

```python
# Handling heteroscedasticity: When the variability of residuals is not constant across all levels of the independent variables
# Using weights to handle heteroscedasticity by assigning lower weights to larger residuals
weights = 1 / np.sqrt(np.abs(model_sm_train.resid))

# Fitting a Weighted Least Squares (WLS) regression model on the training data
# WLS: A method that assigns different weights to each data point to handle heteroscedasticity
model_wls = sm.WLS(y_train, X_with_const_train, weights=weights).fit()

# Printing the summary of the WLS regression model on training data
print(model_wls.summary())

# Applying the same weights to the test set
weights_test = 1 / np.sqrt(np.abs(model_sm_test.resid))

# Fitting a WLS regression model on the testing data with the same weights
model_wls_test = sm.WLS(y_test, X_with_const_test, weights=weights_test).fit()

# Printing the summary of the WLS regression model on testing data
print(model_wls_test.summary())
```

```python
geolocator = Nominatim(user_agent="MyApp")
```

```python
from time import sleep

# Initialize lists to store latitude and longitude
latitudes = []
longitudes = []

# Initialize cache
cache = {}

# Function to get location from cache or geocode
def get_location(place):
    if place in cache:
        return cache[place]
    location = geolocator.geocode(place)
    cache[place] = location
    sleep(1)  # Sleep to avoid hitting the API rate limit
    return location

for index, row in final.iterrows():
    location = None
    
    # Attempt to geocode using city
    if pd.notna(row['City']):
        location = get_location(row['City'])

    # If city is not found, attempt to geocode using state
    if location is None and pd.notna(row['State']):
        location = get_location(row['State'])

    # If state is not found, attempt to geocode using country
    if location is None and pd.notna(row['Country']):
        location = get_location(row['Country'])

    # Append latitude and longitude if location is found, otherwise append None
    if location:
        latitudes.append(location.latitude)
        longitudes.append(location.longitude)
    else:
        latitudes.append(None)
        longitudes.append(None)

# Add the latitude and longitude to the DataFrame
final['Latitude'] = latitudes
final['Longitude'] = longitudes

final.isnull().sum()
final.to_excel('final.xlsx', index=False)
```
